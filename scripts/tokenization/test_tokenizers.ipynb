```python
slurp_text_lines = []
with open("Tokenizer_SLURP/test_transcript_words_custom_slurp.txt") as f:
    for line in f:
        slurp_text_lines.append(line.strip())

libri_text_lines = []
with open("Tokenizer_SLURP/LIBRI_test_transcript_dev_clean.txt") as f:
    for line in f:
        libri_text_lines.append(line.strip())
        
```

```python
libri_trained_tokenizer_path = "Tokenizer_SLURP/models_LIBRI_2000/libri_bpe_2000.model"
slurp_trained_tokenizer_path = "Tokenizer_SLURP/model_vocab_updated_2100/slurp_updated_2100.model"
```

```python
import sentencepiece as spm

libri_trained_tokenizer = spm.SentencePieceProcessor(libri_trained_tokenizer_path)
slurp_trained_tokenizer = spm.SentencePieceProcessor(slurp_trained_tokenizer_path)
```

```python
import re

def remove_tags(text):
    return re.sub(r'<[^>]*>', '', text)
```

```python
from tqdm.auto import tqdm

libri_trained_on_libri_text = []
slurp_trained_on_libri_text = []
for line in tqdm(libri_text_lines, desc="Processing libri test data"):
    libri_trained_on_libri_text.append(libri_trained_tokenizer.encode(line.upper(), out_type=str))
    slurp_trained_on_libri_text.append(slurp_trained_tokenizer.encode(line.lower(), out_type=str))

libri_trained_on_slurp_text = []
slurp_trained_on_slurp_text = []
for line in tqdm(slurp_text_lines, desc="Processing slurp test data"):
    line = remove_tags(line)
    libri_trained_on_slurp_text.append(libri_trained_tokenizer.encode(line.upper(), out_type=str))
    slurp_trained_on_slurp_text.append(slurp_trained_tokenizer.encode(line.lower(), out_type=str))
```

```python
# libri text
print(f"Libri trained tokenizer on libri test data has {str(sum(map(len, libri_trained_on_libri_text)))} tokens")
print(f"Slurp trained tokenizer on libri test data has {str(sum(map(len, slurp_trained_on_libri_text)))} tokens")
```

```python
# slurp text
print(f"Libri trained tokenizer on slurp test data has {str(sum(map(len, libri_trained_on_slurp_text)))} tokens")
print(f"Slurp trained tokenizer on slurp test data has {str(sum(map(len, slurp_trained_on_slurp_text)))} tokens")
```
